```python
import networkx as nx

# Create a directed graph
G = nx.DiGraph()
G.add_edges_from([(0, 1), (0, 2), (1, 3), (2, 3), (2, 4)])

# Dictionary to store incoming nodes for each node
incoming_nodes = []

# Iterate over all nodes
for i, node in enumerate(G.nodes()):
    # Get the nodes that have edges pointing to the current node
    incoming_nodes.append([n for n in G.predecessors(node)])

# Print incoming nodes for each node
for i, nodes in enumerate(incoming_nodes):
    print(f'Node {i}: Incoming nodes {nodes}')
```

## Heatmap Total Clustering

```python
import seaborn as sns
# Set the range of z values
z_range = range(z_max)

# Plotting 25 2D heatmaps of x, y for each depth of z
fig, axes = plt.subplots(5, 5, figsize=(20, 20))
axes = axes.flatten()

for i, z in enumerate(z_range):
    ax = axes[i]
    subset = df[df['z'] == z]
    heatmap_data = pd.crosstab(subset['y'], subset['x'])
    sns.heatmap(heatmap_data, ax=ax, cbar=False)
    ax.set_title(f'Heatmap for z = {z}')
    ax.set_xlabel('x')
    ax.set_ylabel('y')

plt.tight_layout()
plt.show()
```

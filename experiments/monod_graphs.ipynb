```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()

xs = np.linspace(0, 0.1, 100)
ys = (0.002*xs)/(0.005+xs)

plt.plot(xs, ys)
plt.title("Monod equation for $\mu_{max}=0.002$, $K_s=0.005$, and $S=[0,0.1]$")
ax.set_xlabel("S in g")
ax.set_ylabel("$\mu$ in g")
plt.axvline(x = 0.005, c='black')
plt.axvline(x = 0.08, c='black')
plt.text(-0.0045,0.0014,'$S \ll K_s$')
plt.text(0.035,0.0014,'Center Region')
plt.text(0.085,0.0014,'$S \gg K_s$')

plt.savefig("4-monod-regimes.pdf", format="pdf", bbox_inches="tight")
plt.show()
```

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()

xs = np.linspace(0, 0.1, 100)

ys = (0.002*xs)/(0.05+xs)
plt.plot(xs, ys, label="$K_s=0.05$")

ys = (0.002*xs)/(0.01+xs)
plt.plot(xs, ys, label="$K_s=0.01$")

ys = (0.002*xs)/(0.005+xs)
plt.plot(xs, ys, label="$K_s=0.005$")

ys = (0.002*xs)/(0.001+xs)
plt.plot(xs, ys, label="$K_s=0.001$")

ax.set_ylabel("$\mu$ in g")
ax.set_xlabel("S in g")

plt.title("Monod equation for $\mu_{max}=0.002$, $S=[0,0.1]$, and varying $K_s$")
plt.legend()

plt.savefig("4-hs-effect.pdf", format="pdf", bbox_inches="tight")
plt.show()
```

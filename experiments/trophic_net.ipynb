```python
import networkx as nx
import matplotlib
import json

matplotlib.rcParams['path.simplify'] = False
```

```python
G = nx.DiGraph()

G.add_nodes_from(
    [
        (0, {"organism": "bacteria"}),
        (1, {"organism": "fungi"}),
        #(2, {"organism": "amoebae"}),
        #(3, {"organism": "flagellates"}),
        (2, {"organism": "root-feeding nematodes"}),
        (3, {"organism": "bacterivorous nematodes"}),
        (4, {"organism": "fungivorous nematodes"}),
        (5, {"organism": "omnivorous nematodes"}),
        (6, {"organism": "fungivorous mites"}),
        (7, {"organism": "omnivorous mites"}),
        (8, {"organism": "collembolans"}),
        (9, {"organism": "som"}),
    ]
)
G.add_edge(0, 3)
G.add_edge(0, 5)

G.add_edge(1, 4)
G.add_edge(1, 6)
G.add_edge(1, 8)

G.add_edge(2, 5)
G.add_edge(2, 7)

G.add_edge(3, 5)
G.add_edge(3, 7)

G.add_edge(4, 5)
G.add_edge(4, 7)

G.add_edge(5, 7)

G.add_edge(6, 7)

G.add_edge(8, 7)

G.add_edge(9, 0)
G.add_edge(9, 1)
G.add_edge(9, 2)

json.dumps(nx.node_link_data(G))
```

```python
options = {
    "node_color": "black",
    "node_size": 100,
    "width": 3,
}
nx.draw(G, with_labels=True)
```

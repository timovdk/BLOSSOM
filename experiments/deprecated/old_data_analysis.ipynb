```python
import pandas as pd
from sklearn.metrics import mean_absolute_error, median_absolute_error, r2_score, accuracy_score, precision_score, recall_score
```

# Abundance

## Sample

```python
baseline = pd.read_csv('outputs_st/baseline_abundances.csv')
estimate_reg = pd.read_csv('outputs_st/estimated_abundances_reg.csv')
estimate_w = pd.read_csv('outputs_st/estimated_abundances_w.csv')
rs = [1, 2, 3, 4, 5]

metrics_reg = pd.DataFrame(columns=['filename', 'r', 'sample_id', 'mae', 'mdae', 'r2'])
metrics_w = pd.DataFrame(columns=['filename', 'r', 'sample_id', 'mae', 'mdae', 'r2'])
for file in baseline['filename'].unique():
    bl = baseline[baseline['filename'] == file]
    bl = bl[['0', '1', '2', '3', '4', '5', '6', '7', '8']].iloc[0].values
    
    es_r = estimate_reg[estimate_reg['filename'] == file]
    es_w = estimate_w[estimate_w['filename'] == file]
    
    for r in rs:
        es_r_r = es_r[es_r['r'] == r]
        es_w_r = es_w[es_w['r'] == r]
        
        for i in range(len(es_r_r['sample_id'].unique())):
            es_r_ab = es_r_r[es_r_r['sample_id'] == i][['0', '1', '2', '3', '4', '5', '6', '7', '8']].iloc[0].values
            temp = {'filename': file, 'r': r, 'sample_id': i, 'mae': mean_absolute_error(bl, es_r_ab), 'mdae': median_absolute_error(bl, es_r_ab), 'r2': r2_score(bl, es_r_ab)}
            metrics_reg = pd.concat([metrics_reg, pd.DataFrame([temp])])
            
        for i in range(len(es_w_r['sample_id'].unique())):
            es_w_ab = es_w_r[es_w_r['sample_id'] == i][['0', '1', '2', '3', '4', '5', '6', '7', '8']].iloc[0].values
            temp = {'filename': file, 'r': r, 'sample_id': i, 'mae': mean_absolute_error(bl, es_w_ab), 'mdae': median_absolute_error(bl, es_w_ab), 'r2': r2_score(bl, es_w_ab)}
            metrics_w = pd.concat([metrics_w, pd.DataFrame([temp])])

metrics_w.to_csv('abundances_sample_w.csv', index=False)
metrics_reg.to_csv('abundances_sample_reg.csv', index=False)
```

```python
baseline = pd.read_csv('outputs/baseline_diversity_indices.csv')
estimate_reg = pd.read_csv('outputs/estimated_diversity_indices_sample_reg.csv')
estimate_w = pd.read_csv('outputs/estimated_diversity_indices_sample_w.csv')
rs = [1, 2, 3, 4, 5]

metrics_reg = pd.DataFrame(columns=['filename', 'r', 'sample_id', 'mae_sh', 'mae_si', 'mdae_sh', 'mdae_si'])
metrics_w = pd.DataFrame(columns=['filename', 'r', 'sample_id', 'mae_sh', 'mae_si', 'mdae_sh', 'mdae_si'])
for file in baseline['filename'].unique():
    bl_f = baseline[baseline['filename'] == file]
    bl_sh = bl_f['shannon'].values[0]
    bl_si = bl_f['simpson'].values[0]
    es_r = estimate_reg[estimate_reg['filename'] == file]
    es_w = estimate_w[estimate_w['filename'] == file]

    for sample_id in range(len(es_r['sample_id'].unique())):
        es_r_s = es_r[es_r['sample_id'] == sample_id]
        
        for r in rs:
            es_r_r = es_r_s[es_r_s['diameter'] == r]
            sh = es_r_r['shannon'].values[0]
            si = es_r_r['simpson'].values[0]
            temp = {'filename': file, 'r': r, 'sample_id': sample_id, 'mae_sh': mean_absolute_error([bl_sh], [sh]), 'mae_si': mean_absolute_error([bl_si], [si]), 'mdae_sh': median_absolute_error([bl_sh], [sh]), 'mdae_si': median_absolute_error([bl_si], [si])}
            metrics_reg = pd.concat([metrics_reg, pd.DataFrame([temp])])
    
    for sample_id in range(len(es_w['sample_id'].unique())):
        es_w_s = es_w[es_w['sample_id'] == sample_id]
        
        for r in rs:
            es_w_r = es_w_s[es_w_s['diameter'] == r]
            sh = es_w_r['shannon'].values[0]
            si = es_w_r['simpson'].values[0]
            temp = {'filename': file, 'r': r, 'sample_id': sample_id, 'mae_sh': mean_absolute_error([bl_sh], [sh]), 'mae_si': mean_absolute_error([bl_si], [si]), 'mdae_sh': median_absolute_error([bl_sh], [sh]), 'mdae_si': median_absolute_error([bl_si], [si])}
            metrics_w = pd.concat([metrics_w, pd.DataFrame([temp])])
                
metrics_w.to_csv('diversity_indices_sample_w.csv', index=False)
metrics_reg.to_csv('diversity_indices_sample_reg.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import colormaps
from matplotlib.gridspec import GridSpec

filenames = ['random_1', 'random_2', 'random_3', 'random_4', 'random_5', 'random_6', 'random_7', 'random_8', 'random_9', 'random_10', 'random_11', 'random_12', 'random_13', 'random_14', 'random_15', 'random_16', 'random_17', 'random_18', 'random_19', 'random_20', 'clustered_1', 'clustered_2', 'clustered_3', 'clustered_4', 'clustered_5', 'clustered_6', 'clustered_7', 'clustered_8', 'clustered_9', 'clustered_10', 'clustered_11', 'clustered_12', 'clustered_13', 'clustered_14', 'clustered_15', 'clustered_16', 'clustered_17', 'clustered_18', 'clustered_19', 'clustered_20']
path = "../blossom/hpc/outputs/"
sample_times = [252, 415, 310, 192, 241, 308, 338, 231, 241, 233, 241, 330, 246, 246, 161, 327, 158, 218, 291, 431, 206, 185, 307, 258, 287, 244, 335, 291, 151, 206, 167, 356, 208, 224, 271, 181, 179, 279, 305, 246]
organism_group_labels = ["Bacteria", "Fungi", "Root-feeding Nematodes", "Bacterivorous Nematodes", "Fungivorous Nematodes", "Omnivorous Nematodes", "Fungivorous Mites", "Omnivorous Mites", "Collembolans"]
x_max = 400
y_max = 400

def rand_jitter(arr):
    if(len(arr)):
        stdev = .001 * (max(arr) - min(arr))
        return arr + np.random.randn(len(arr)) * stdev
    return arr


df = pd.read_csv(path + filename + ".csv")
n = len(df["type"].unique())
colors = colormaps['tab10'].colors

counts_per_type = df.value_counts(['type', 'tick'])
```

```python
fig = plt.figure(figsize=(11.69,20))

gs = GridSpec(4, 2)
ax1 = fig.add_subplot(gs[0])
ax2 = fig.add_subplot(gs[1])
ax3 = fig.add_subplot(gs[2])
ax4 = fig.add_subplot(gs[3])
ax5 = fig.add_subplot(gs[4])
ax6 = fig.add_subplot(gs[5])
ax7 = fig.add_subplot(gs[6])
ax8 = fig.add_subplot(gs[7])

for i in df['type'].unique():
    ax1.plot(
        range(len(df["tick"].unique())),
        counts_per_type[i].reindex(range(len(df['tick'].unique())), fill_value=0).sort_index().to_list(),
        label=organism_group_labels[i],
        color=colors[i]
    )
ax1.set_xlim(0, df["tick"].max())
ax1.set_ylim(0, 150000)
ax1.set_xlabel("Ticks")
ax1.set_ylabel("Agent Count")
ax1.set_title(f'{filename}, Counts')
ax1.legend()

df1 = df[df["tick"] == 0]
for t in np.unique(df1["type"]):
    subset_type = df1[df1['type'] == t]
    ax2.scatter(rand_jitter(subset_type['x']), rand_jitter(subset_type['y']), color=colors[t], label=t, s=1, marker='s')
ax2.set_title('Tick: 0')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

df1 = df[df["tick"] == 100]
for t in np.unique(df1["type"]):
    subset_type = df1[df1['type'] == t]
    ax3.scatter(rand_jitter(subset_type['x']), rand_jitter(subset_type['y']), color=colors[t], label=t, s=1, marker='s')
ax3.set_title('Tick: 100')
ax3.set_xlabel('X')
ax3.set_ylabel('Y')

df1 = df[df["tick"] == 200]
for t in np.unique(df1["type"]):
    subset_type = df1[df1['type'] == t]
    ax4.scatter(rand_jitter(subset_type['x']), rand_jitter(subset_type['y']), color=colors[t], label=t, s=1, marker='s')
ax4.set_title('Tick: 200')
ax4.set_xlabel('X')
ax4.set_ylabel('Y')

df1 = df[df["tick"] == 300]
for t in np.unique(df1["type"]):
    subset_type = df1[df1['type'] == t]
    ax5.scatter(rand_jitter(subset_type['x']), rand_jitter(subset_type['y']), color=colors[t], label=t, s=1, marker='s')
ax5.set_title('Tick: 300')
ax5.set_xlabel('X')
ax5.set_ylabel('Y')

df1 = df[df["tick"] == 400]
for t in np.unique(df1["type"]):
    subset_type = df1[df1['type'] == t]
    ax6.scatter(rand_jitter(subset_type['x']), rand_jitter(subset_type['y']), color=colors[t], label=t, s=1, marker='s')
ax6.set_title('Tick: 400')
ax6.set_xlabel('X')
ax6.set_ylabel('Y')

df1 = df[df["tick"] == 500]
for t in np.unique(df1["type"]):
    subset_type = df1[df1['type'] == t]
    ax7.scatter(rand_jitter(subset_type['x']), rand_jitter(subset_type['y']), color=colors[t], label=t, s=1, marker='s')
ax7.set_title('Tick: 500')
ax7.set_xlabel('X')
ax7.set_ylabel('Y')

df1 = df[df["tick"] == 600]
for t in np.unique(df1["type"]):
    subset_type = df1[df1['type'] == t]
    ax8.scatter(rand_jitter(subset_type['x']), rand_jitter(subset_type['y']), color=colors[t], label=t, s=1, marker='s')
ax8.set_title('Tick: 600')
ax8.set_xlabel('X')
ax8.set_ylabel('Y')

# Adjust layout to make sure plots are neatly arranged
plt.tight_layout()
plt.savefig("app-hm-" + filename + ".pdf", format="pdf", bbox_inches="tight")
plt.savefig("app-hm-" + filename + ".png", format="png", bbox_inches="tight")
plt.show()
```
